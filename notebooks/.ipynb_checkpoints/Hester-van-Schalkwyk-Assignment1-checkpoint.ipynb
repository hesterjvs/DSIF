# Import packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("../data/2-intermediate/df_out_dsif2.csv")

# Display basic info
df.info()
df.head()
```

# <span style="color:TEAL">**---- Insight 1: Relationship between Interest Rate and Loan Default ----**</span>  

```python
# Convert interest rate to numeric
df['int_rate'] = df['int_rate'].str.rstrip('%').astype(float)

# Check default rate by interest rate bins
df['defaulted'] = df['loan_status'].apply(lambda x: 1 if 'Charged Off' in str(x) else 0)
df['int_rate_bin'] = pd.cut(df['int_rate'], bins=[0, 10, 15, 20, 30], labels=['Low', 'Medium', 'High', 'Very High'])

# Count number of Charged Off loans
charged_off_count = df['defaulted'].sum()
total_loans = len(df)
print(f"Number of Charged Off loans: {charged_off_count} ({(charged_off_count / total_loans) * 100:.2f}%)")

# Plot default rates by interest rate bins
plt.figure(figsize=(8,5))
sns.barplot(x=df.groupby('int_rate_bin', observed=True)['defaulted'].mean().index, 
            y=df.groupby('int_rate_bin', observed=True)['defaulted'].mean().values, 
            hue=df.groupby('int_rate_bin', observed=True)['defaulted'].mean().index,
            dodge=False,
            palette='coolwarm')
plt.title("Default Rate by Interest Rate Category")
plt.ylabel("Default Rate")
plt.xlabel("Interest Rate Category")
handles, labels = plt.gca().get_legend_handles_labels()
if labels:
    plt.legend(handles, labels)
plt.show()
```

### Business Insight 1: 
Higher interest rates correlate with higher default rates. Understanding this trend helps in pricing loans appropriately to balance profitability and risk.
### Action: 
Adjust lending criteria to mitigate risk.
### Measurement: 
Track loan performance in different interest rate brackets. Evaluate Long-Term Borrower Behavior â€“ Assess if borrowers with slightly lower interest rates (in a risk-based model) default less often and stay with the bank longer, increasing customer lifetime value.

# <span style="color:TEAL">**---- Insight 2: Employment Length vs Loan Default ----**</span>

```python
# Extract numeric part
df['emp_length_num'] = df['emp_length'].str.extract(r'(\d+)').astype(float)  

def categorize_emp_length(x):
    if pd.isnull(x):
        return 'Unknown'
    elif x <= 2:
        return '0-2 years'
    elif x <= 5:
        return '3-5 years'
    elif x <= 10:
        return '6-10 years'
    else:
        return '10+ years'

df['emp_length_category'] = df['emp_length_num'].apply(categorize_emp_length)


# Plot default rates by employment length
plt.figure(figsize=(8,5))
sns.barplot(x=df.groupby('emp_length_category', observed=True)['defaulted'].mean().index, 
            y=df.groupby('emp_length_category', observed=True)['defaulted'].mean().values, 
            hue=df.groupby('emp_length_category', observed=True)['defaulted'].mean().index,
            dodge=False,
            palette='coolwarm')
plt.title("Default Rate by Employment Length")
plt.ylabel("Default Rate")
plt.xlabel("Employment Length Category")
handles, labels = plt.gca().get_legend_handles_labels()
if labels:
    plt.legend(handles, labels)
plt.show()

# Boxplot to show distribution of employment length for defaulters vs non-defaulters
plt.figure(figsize=(8,5))
sns.boxplot(x=df['defaulted'], y=df['emp_length_num'], hue=df['defaulted'], palette='coolwarm', legend=False)
plt.title("Distribution of Employment Length by Default Status")
plt.ylabel("Employment Length (Years)")
plt.xlabel("Default Status (0 = No, 1 = Yes)")
plt.show()
```

### Business Insight 2: 
 The correlation between employment length and default rates appears weak, suggesting other factors may play a larger role in loan repayment.
#### Action: 
 Instead of using employment length alone, lenders should incorporate additional factors such as income stability, credit history, and debt-to-income ratio into risk assessment models.
#### Measurement: 
 Compare loan performance by employment length in combination with other risk factors, testing different weights in risk models to optimize predictive accuracy.

# <span style="color:TEAL">**---- Insight 3: Loan Purpose vs Default Rate ----**</span>

```python
# Group by loan purpose and calculate default rates
purpose_default_rates = df.groupby('purpose')['defaulted'].mean().sort_values(ascending=False)

# Plot default rates by loan purpose
plt.figure(figsize=(10, 6))
sns.barplot(x=purpose_default_rates.index, y=purpose_default_rates.values, hue=purpose_default_rates.index, dodge=False, palette='coolwarm', legend=False)
plt.title("Default Rate by Loan Purpose")
plt.ylabel("Default Rate")
plt.xlabel("Loan Purpose")
plt.xticks(rotation=45)
plt.show()
```

### Business Insight 3: 
 Different loan purposes have varying default rates, with some categories significantly riskier than others. For instance, loans for discretionary expenses like vacations or luxury purchases may show higher default rates compared to essential needs like home improvement or debt consolidation.
#### Action: 
 Lenders should consider adjusting interest rates or approval criteria based on loan purpose risk. Higher-risk categories may require stricter credit evaluations or collateral requirements.
#### Measurement: 
 Monitor loan performance across different purposes over time and refine lending criteria accordingly, reducing defaults while maintaining a competitive lending portfolio.

# <span style="color:TEAL">**---- Insight 4: Annual Income vs. Loan Default ----**</span>

```python
# Create income bins
df['income_bin'] = pd.cut(df['annual_inc'], bins=[0, 30000, 60000, 100000, float('inf')], 
                           labels=['Low', 'Medium', 'High', 'Very High'])

# Plot default rates by income bins
plt.figure(figsize=(8, 5))
sns.barplot(x=df.groupby('income_bin', observed=True)['defaulted'].mean().index, 
            y=df.groupby('income_bin', observed=True)['defaulted'].mean().values, 
            hue=df.groupby('income_bin', observed=True)['defaulted'].mean().index,
            dodge=False,
            palette='coolwarm')
plt.title("Default Rate by Income Category")
plt.ylabel("Default Rate")
plt.xlabel("Income Category")
handles, labels = plt.gca().get_legend_handles_labels()
if labels:
    plt.legend(handles, labels)
plt.show()
```

### Business Insight 4:
Higher-income borrowers tend to default less frequently, whereas lower-income groups show higher default rates.
#### Action:
Lenders should consider adjusting credit evaluations based on income levels, potentially requiring additional financial verification for lower-income borrowers.
#### Measurement:
Track default rates across income categories over time and assess how additional income verification impacts loan performance.

# <span style="color:TEAL">**---- Insight 5: Exploring correlations between key numerical features ----**</span>

```python
# Correlation Analysis: 
plt.figure(figsize=(10, 6))
sns.heatmap(df[['annual_inc', 'loan_amnt', 'int_rate', 'defaulted']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### Business Insight 5:
- **Loan Amount and Interest Rate:** Positive correlation suggests higher loan amounts are associated with higher interest rates, possibly indicating increased risk pricing.
- **Annual Income and Loan Amount:** Weak correlation, meaning higher-income individuals do not necessarily take larger loans.
- **Interest Rate and Default Rate:** Moderate correlation suggests that higher interest rates might be a factor in loan defaults.
#### Action:
Refine lending strategies by balancing risk pricing with affordability analysis to prevent high-interest burdens leading to defaults.
#### Measurement:
Monitor how adjustments in interest rate structures impact loan performance and overall portfolio risk.
