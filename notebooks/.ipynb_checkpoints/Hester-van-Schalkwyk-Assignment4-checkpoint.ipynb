## Step 1: Load and Explore Data

Objective:
- Load the customer credit card dataset.
- Understand its structure and check for missing values.
- Identify numerical features for clustering.

Assignment Relevance:
- This step ensures we clean and prepare the dataset for clustering.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Load the dataset
df = pd.read_csv('../data/1-raw/dsif8-assignment/Customer_Data.csv')

# Display basic info
display(df.head())
display(df.info())

# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())

# Drop missing values if necessary
df.dropna(inplace=True)
```

## Step 2: Identify and Preprocess Numeric Features

Objective:
- Identify numerical columns.
- Standardize numerical data to ensure fair clustering.

Assignment Relevance:
- Standardizing data allows clustering algorithms to work efficiently and accurately.

```python
# Identify non-numeric columns
print("Column data types:")
print(df.dtypes)

# Select only numeric columns for clustering
numerical_cols = df.select_dtypes(include=['number']).columns
df_numeric = df[numerical_cols]

# Standardize the numeric data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numeric)
```

## Step 3: Dimensionality Reduction using PCA

Objective:
- Reduce dataset complexity while preserving essential patterns.
- Enable visualization of high-dimensional data.

Assignment Relevance:
- PCA helps simplify data and allows for better interpretation of clusters.

```python
# Apply PCA for visualization (2 components)
pca = PCA(n_components=2)
pca_components = pca.fit_transform(df_scaled)

# Convert PCA results into a DataFrame
pca_df = pd.DataFrame(data=pca_components, columns=['PC1', 'PC2'])
```

## Step 4: Determine Optimal Clusters

Objective:
- Use the Elbow Method to determine the best number of clusters.

Assignment Relevance:
- Selecting an appropriate cluster number ensures meaningful segmentation.

```python
# Determine the optimal number of clusters using the Elbow Method
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(8, 4))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()
```

## Step 5: Apply K-Means Clustering

Objective:
- Assign customers to clusters using K-Means.
- Evaluate clustering performance using silhouette score.

Assignment Relevance:
- Clustering helps segment customers based on spending behaviour.

```python
# Choose the optimal number of clusters (from elbow method, let's assume k=4)
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(df_scaled)

# Compute silhouette score
sil_score = silhouette_score(df_scaled, df['Cluster'])
print(f"Silhouette Score: {sil_score:.4f}")
```

## Step 6: Visualizing Clusters using PCA

Objective:
- Visualize customer clusters in two dimensions.

Assignment Relevance:
- Helps understand the separa

```python
# Add cluster labels to the PCA DataFrame
pca_df['Cluster'] = df['Cluster']

# Plot the clusters based on PCA components
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=pca_df, palette='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Customer Segments based on PCA')
plt.legend()
plt.show()
```

## Step 7: Interpret Clusters

Objective:
- Analyze the characteristics of each cluster.
- Count the number of observations in each cluster.

Assignment Relevance:
- Helps businesses understand customer behavior and tailor marketing strategies accordingly.


```python
# Analyze cluster characteristics
cluster_analysis = df.groupby('Cluster')[numerical_cols].mean()
display(cluster_analysis)
```

```python
# Count number of observations per cluster
cluster_counts = df['Cluster'].value_counts()
print("Number of observations per cluster:")
print(cluster_counts)
```

## Findings and Insights

Cluster Characteristics: The `cluster_analysis` DataFrame provides the average values of features for each cluster, helping to understand the distinct characteristics of each customer segment.​

Visualization: The PCA scatter plot visualizes the distribution of customers across different segments, illustrating how distinct or overlapping the clusters are. In this case they overlap, and are not very distinct.​

By following this approach, you can effectively segment the credit card customer base, enabling targeted marketing and informed strategic decisions.
